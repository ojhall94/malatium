```python
import pandas as pd
import numpy as np
from model import mix
```

```python
df = pd.read_csv('../data/data_for_guy.csv', index_col=0)
```

```python
df.head()
df.loc[df.hrclass == 'MS']
```

```python
stars = []
for idx, row in df.iterrows():
    mass = [row.modmass, max([row.upmodmass, row.lomodmass])]
    age = [row.age, max([row.upage, row.loage])]
    teff = [row.Teff, row.eTeff]
    prot = [row.P, max([row.l_P, row.u_P])]
    d = {'ID' : str(row.KIC),
        'mass' : mass,
        'teff' : teff,
        'age' : age,
         'prot' : prot}
    stars.append(d)
```

```python
stars[0]
```

```python
%%time
mix = mix()
mix.run_one_star(stars[0])
```
