```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

sns.set_palette('colorblind')
sns.set_context('poster')
```

```python
ati = pd.read_csv('../data/atium.csv', index_col=0)
ati
```

```python
ati.nus
```

```python
bro = pd.read_csv('../data/bronze.csv', index_col=0)
bro
```

```python
import matplotlib as mpl
norm = mpl.colors.Normalize(vmin=ati.P.min(),vmax=ati.P.max())
cmap = plt.cm.viridis(norm(ati.P.values))
```

```python
fig, axes = plt.subplots(2,1, figsize=(20, 8))
x = np.arange(len(ati))
ati['gammaflag'] = np.zeros(len(ati))
for idx, kic in enumerate(ati.KIC.values[:95]):
    if idx > 50:
        ax = axes[1]
    else: ax = axes[0]
    
    b = bro[bro.KIC == str(kic)]
    c = ax.fill_between([x[idx]-0.25, x[idx]+0.25], [b.g.min(), b.g.min()], [b.g.max(), b.g.max()],
                   alpha=.5, color=cmap[idx])
    
    ax.scatter(x[idx], ati.loc[idx, 'nus']/2, marker=',', c='k', s=50)
    ax.set_yscale('log')
    
    if ati.loc[idx, 'nus']/2 < b.g.max():
        ati.loc[idx, 'gammaflag'] = 1
    
fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap='viridis'), ax=axes, pad=-0.001, alpha=.5, 
             label='Seismic Rotation [days]')
axes[0].set_title(r'Is $\nu_{\rm s} > 2\Gamma$ ? | Bars: range of linewidth per star. | Squares: $0.5\times\nu_{\rm s}$')
axes[1].set_ylabel('Splitting or Linewidth')
axes[0].set_ylabel(r'[$\mu\rm{Hz}$]')


plt.show()
```
