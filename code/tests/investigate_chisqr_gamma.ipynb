```python
import numpy as np
import pylab as plt
```

```python
from scipy.stats import chi2
def chi_square_2dof(x):
    return chi2.pdf(x, 2)
```

```python
from scipy.special import gamma as gammafn
def gamma(x, beta):
    return (beta/gammafn(1)) * np.exp(-x*beta)
```

```python
x = np.linspace(0, 5, 1000)
m = np.sin(x)
# plt.plot(x, m, label='model')
fig = plt.figure(figsize=(16,12))
plt.plot(x, chi_square_2dof(x), label='chisqr', linewidth=5)
plt.plot(x, chi_square_2dof(x)*m, label='chisqr*m', linewidth=5)
plt.plot(x, gamma(x, m), label='gamma, beta=m', linewidth=5)
plt.plot(x, gamma(x, m/2), label='gamma, beta=m/2', linewidth=5)
plt.plot(x, gamma(x, abs(1/m)), label='gamma, beta=1/m', linewidth=5)
plt.ylim(-1, 1)
plt.legend(fontsize=25)
plt.show()
```
