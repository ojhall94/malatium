```python
import lightkurve as lk
import numpy as np
```

```python
x = np.linspace(0., 1000., 1000.)
y = np.random.randn(len(x))*np.sin(x)*10
import pylab as plt
plt.plot(x,y)
```

```python
lc = lk.LightCurve(x, y)
lc.plot()
```

```python
pg = lc.to_periodogram(normalization='psd')
pg.plot()
```

```python
print(type(lc))
print(type(pg))
```

```python
f, p = pg.frequency.value, pg.power.value
```

```python
from astropy import units as u
pg = lk.periodogram.Periodogram(f*u.microhertz, u.Quantity(p, None))
pg.plot()
```
