```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
m = 3
b = 5
y = x * m + b
plt.scatter(x, y)
```

```python
T = tf.float64
m_tens = tf.Variable(m, dtype=T)
b_tens = tf.Variable(b, dtype=T)
x_tens = tf.constant(x, dtype=T)
y_tens = tf.constant(y, dtype=T)
```

```python
def tf_model(x, m, b):
    return x*m + b
```

```python
model = tf_model(x_tens, m_tens, b_tens)
with tf.Session() as session:
    session.run(tf.global_variables_initializer())
    mod = session.run(model)
    plt.scatter(x, y)
    plt.plot(x, mod)
    plt.show()
```
