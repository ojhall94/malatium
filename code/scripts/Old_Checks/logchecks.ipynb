```python
import seaborn as sns
import pylab as plt
import numpy as np
```

```python
A = np.random.randn(1000) + 4
A = np.log10(A)
sns.distplot(A)
```

```python
a, b, c = np.percentile(A, [16, 50, 84])
sns.distplot(A)
plt.axvline(a, ls='--')
plt.axvline(b)
plt.axvline(c, ls='--')
```

Now in log

```python
B = np.log10(A)
sns.distplot(B)
```

```python
j, h, k = np.percentile(B, [16, 50, 84])
sns.distplot(B)
plt.axvline(j, ls='--')
plt.axvline(h)
plt.axvline(k, ls='--')
```

```python
print(f'The median of A is: {b:.2f}')
print(f'The median of B is: {h:.2f}, and 10^{h:.2f} is: {10**h:.2f}')
```
