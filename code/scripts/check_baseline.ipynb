```python
import numpy as np
```

```python
from astropy.io import ascii
```

```python
import glob
```

```python
flist = glob.glob('../../data/*.pow')
```

```python
f = ascii.read(flist[0])
```

```python
fs = np.median(np.diff(f['col1']))
```

```python
import pandas as pd
bl = pd.DataFrame()
bl['f'] = flist
bl['bl'] = np.nan
```

```python
bl.bl.iloc[0]
```

```python
for idx, file in enumerate(flist):
    f = ascii.read(file)
    fs = np.median(np.diff(f['col1']))
    bl.bl.iloc[idx] = np.round((1/(fs*1e-6)) / 3600 / 24 / 365.25,2)
```

```python
bl
bl.to_csv('baselines.csv')
```

```python
bl.sort_values(by='bl')
```
